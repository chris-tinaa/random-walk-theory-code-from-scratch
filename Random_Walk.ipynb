```
import random
import matplotlib.pyplot as plt 

n_particle = 10
n_iteration = 100
x_pos = []
y_pos = []

# create empty 2D array
for i in range(n_particle):
  x_pos.append([0])
  y_pos.append([0])

for i in range(n_iteration):
  if i == 0 :
    continue

  for j in range(n_particle):
    step = random.choice(['U', 'D', 'R', 'L'])
    # Up : y=y+1 dan x=x
    if step == 'U':
      y_pos[j].append(y_pos[j][i-1] + 1)
      x_pos[j].append(x_pos[j][i-1])
    # Down : y=y-1 dan x=x
    elif step == 'D':
      y_pos[j].append(y_pos[j][i-1] - 1)
      x_pos[j].append(x_pos[j][i-1])
    # Right : y=y dan x=x+1
    elif step == 'R':
      x_pos[j].append(x_pos[j][i-1] + 1)
      y_pos[j].append(y_pos[j][i-1])
    # Left : y=y dan x=x-1
    elif step == 'L':
      x_pos[j].append(x_pos[j][i-1] - 1)
      y_pos[j].append(y_pos[j][i-1])

# warna tiap partikel
colors = ['#f54242', '#f5a442', '#f58442', '#b3f542', '#42f575', '#42f5d7', '#42b0f5', '#4254f5', '#9e42f5', '#f542cb']

# ambil posisi (x,y) tiap 10 iterasi
s = slice(0, n_iteration, 10)
print('POSISI TIAP 10 ITERASI HINGGA ITERASI KE-N (N = n_iteration = 100)')
for i in range(n_particle): 
  print('Partikel ke-', i+1, ': ')
  print('x = ', x_pos[i][s])
  print('y = ', y_pos[i][s])
  # plotting the points
  plt.plot(x_pos[i][s], y_pos[i][s], color=colors[i], linestyle='solid', linewidth = 1, marker='o', markerfacecolor=colors[i], markersize=2)

# # setting x and y axis range 
# plt.ylim(0,10) 
# plt.xlim(0,10) 
  
# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 
  
# giving a title to my graph 
plt.title('Random Walk') 

# function to show the plot 
plt.show() 


```
